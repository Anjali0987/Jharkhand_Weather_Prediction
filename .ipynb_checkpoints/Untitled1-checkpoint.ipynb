```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
```

```python
clusters = []
for i in range(1,15):
    clusters.append(i)
inertias = []
for k in clusters:
    k = KMeans(n_clusters=k,init='k-means++',random_state=100)
    k.fit(weather)
    inertias.append(k.inertia_)
    
sn.lineplot(x = clusters, y = inertias)
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
mm = MinMaxScaler()
scaled_features = mm.fit_transform(weather)
```

```python
k2 = KMeans(init="random",n_clusters=5,n_init=10,max_iter=500,random_state=100)
```

```python
k2.fit(scaled_features)
```

```python
# The lowest SSE value
k2.inertia_
```

```python
# Final locations of the centroid
c=k2.cluster_centers_
print(c)
```

```python
# Function that creates a DataFrame with a column for Cluster Number
def pd_centers(features, centers):
    col = list(features)
    col.append('LABEL')

    # Zip with a column called 'LABEL' (index)
    Z = [np.append(A, index) for index, A in enumerate(centers)]

    # Convert to pandas data frame for plotting
    P = pd.DataFrame(Z, columns=col)
    P['LABEL'] = P['LABEL'].astype(int)
    return P
```

```python
total_clusters = pd_centers(weather, c)
total_clusters
```

```python
X = weather.iloc[:,:-1]
```

```python
y=weather.iloc[:,-1]
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=100)
```

# Support Vector Machine

```python
from sklearn import svm
```

```python
model=svm.SVC()
```

```python
model.fit(X_train,y_train)
```

```python
ypred=model.predict(X_test)
```

```python
X_test,ypred
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,model.predict(X_test))
```

```python
score=model.score(X_test,y_test)
print(round(score,2))
```

```python
from sklearn.metrics import classification_report 
print(classification_report(y_test,ypred))
```
