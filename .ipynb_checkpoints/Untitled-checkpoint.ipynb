```python
weather=df1.drop(["DATE"],axis=1)
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
```

```python
clusters = [2, 3, 4, 5, 6, 7, 8, 9, 10]
inertias = []
for k in clusters:
    k = KMeans(n_clusters=k,init='k-means++',random_state=42)
    k.fit(weather)
    inertias.append(k.inertia_)

sn.lineplot(x = clusters, y = inertias)
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_features = scaler.fit_transform(weather)
```

```python
scaled_features[:5]
```

```python
k2 = KMeans(init="random",n_clusters=5,n_init=10,max_iter=300,random_state=42)
```

```python
k2.fit(scaled_features)
```

```python
# The lowest SSE value
k2.inertia_
```

```python
# Final locations of the centroid
c=k2.cluster_centers_
print(c)
```

```python
k2.labels_[:5]
```

```python
c = k2.cluster_centers_
c
```

```python
# Function that creates a DataFrame with a column for Cluster Number

def pd_centers(features, centers):
    col = list(features)
    col.append('LABEL')

    # Zip with a column called 'LABEL' (index)
    Z = [np.append(A, index) for index, A in enumerate(centers)]

    # Convert to pandas data frame for plotting
    P = pd.DataFrame(Z, columns=col)
    P['LABEL'] = P['LABEL'].astype(int)
    return P
```

```python
P = pd_centers(scaled_features, c)
P
```
