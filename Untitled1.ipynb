```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
```

```python
from sklearn.preprocessing import MinMaxScaler
mm = MinMaxScaler()
scaled_features = mm.fit_transform(weather)
```

```python
k2 = KMeans(init="random",n_clusters=5,n_init=10,max_iter=500,random_state=100)
```

```python
k2.fit(scaled_features)
```

```python
# The lowest SSE value
k2.inertia_
```

```python
# Final locations of the centroid
c=k2.cluster_centers_
print(c)
```

```python
# Function that creates a DataFrame with a column for Cluster Number
def pd_centers(features, centers):
    col = list(features)
    col.append('LABEL')

    # Zip with a column called 'LABEL' (index)
    Z = [np.append(A, index) for index, A in enumerate(centers)]

    # Convert to pandas data frame for plotting
    P = pd.DataFrame(Z, columns=col)
    P['LABEL'] = P['LABEL'].astype(int)
    return P
```

```python
total_clusters = pd_centers(weather, c)
total_clusters
```

```python
#6 Visualising the clusters
plt.figure(figsize=(16,9))
plt.scatter(m[y_kmeans==0, 0], m[y_kmeans==0, 1], s=100, c='red', label ='Cluster 0 - Cloudy')
plt.scatter(m[y_kmeans==1, 0], m[y_kmeans==1, 1], s=100, c='blue', label ='Cluster 1 - Rainy')
plt.scatter(m[y_kmeans==2, 0], m[y_kmeans==2, 1], s=100, c='green', label ='Cluster 2 - Sunny')
plt.scatter(m[y_kmeans==3, 0], m[y_kmeans==3, 1], s=100, c='cyan', label ='Cluster 3 - Hot')
plt.scatter(m[y_kmeans==4, 0], m[y_kmeans==4, 1], s=100, c='magenta', label ='Cluster 4 - Cold')
#Plot the centroid. This time we're going to use the cluster centres  #attribute that returns here the coordinates of the centroid.
plt.scatter(k2.cluster_centers_[:, 0], k2.cluster_centers_[:, 1], s=300, c='black', label = 'Centroids')
plt.legend()
plt.show()
```

```python
# 3d scatterplot using matplotlib

fig = plt.figure(figsize = (15,15))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(m[y_kmeans == 0,0],m[y_kmeans == 0,1],m[y_kmeans == 0,2], s = 40 , color = 'blue', label = "cluster 0")
ax.scatter(m[y_kmeans == 1,0],m[y_kmeans == 1,1],m[y_kmeans == 1,2], s = 40 , color = 'orange', label = "cluster 1")
ax.scatter(m[y_kmeans == 2,0],m[y_kmeans == 2,1],m[y_kmeans == 2,2], s = 40 , color = 'green', label = "cluster 2")
ax.scatter(m[y_kmeans == 3,0],m[y_kmeans == 3,1],m[y_kmeans == 3,2], s = 40 , color = '#D12B60', label = "cluster 3")
ax.scatter(m[y_kmeans == 4,0],m[y_kmeans == 4,1],m[y_kmeans == 4,2], s = 40 , color = 'purple', label = "cluster 4")
ax.set_xlabel('Maximum Temperature')
ax.set_ylabel('Minimum Temperature')
ax.set_zlabel('Relative Humidity')
ax.legend()
plt.show()
```
